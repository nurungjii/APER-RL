```python
import pandas as pd
import wandb
import os
import glob
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
entity = "github-2-delirium155-washington-university-in-st-louis"  # Replace with your W&B username or team name
project = "ddqn-ER"  # Replace with your project name
run_id = "13s813xg"  # Replace with the specific run ID

# Initialize the API
api = wandb.Api()

# Fetch the specific run
run = api.run(f"{entity}/{project}/{run_id}")

# Fetch the history of logged metrics
history = run.scan_history()
history_df = pd.DataFrame(list(history))

# Specify the metric you want to export
metric = "eval_avg_reward"  # Replace with your desired metric name

# Extract data for the specific metric
metric_data = history[[metric]]

# Export to CSV
metric_data
```

```python
#Asterix ER Calculation
folder_path = 'runs/Asterix/ER/'
x_col = 'Step'
pattern = 'eval_avg_reward'
common_x = np.linspace(0, 50_000_000, num=1000) 
interpolated_values = []

for csv in glob.glob(os.path.join(folder_path, '*.csv')):
    df = pd.read_csv(csv)
    
    max_x = df[x_col].max()
    scaling_factor = 50_000_000 / max_x
    df[x_col] = df[x_col] * scaling_factor

    reward_columns = [col for col in df.columns if col.endswith(pattern)]
    for col in reward_columns:
        valid_data = df[[x_col, col]].dropna().sort_values(by=x_col)
        if not valid_data.empty:
            # Interpolate y-values onto the common x grid
            interpolator = interp1d(
                valid_data[x_col], 
                valid_data[col], 
                bounds_error=False, 
                fill_value="extrapolate"
            )
            interpolated_y = interpolator(common_x)
            interpolated_values.append(interpolated_y)

averaged_y = np.mean(interpolated_values, axis=0)
averaged_asterix_er_df = pd.DataFrame({'x': common_x, 'y': averaged_y})
```

```python
# Apply a rolling mean with window size 10
window = 50
averaged_asterix_er_df['y_smoothed'] = averaged_asterix_er_df['y'].rolling(window=window, center=True).mean()
averaged_asterix_per_df['y_smoothed'] = averaged_asterix_per_df['y'].rolling(window=window, center=True).mean()
averaged_asterix_aper_df['y_smoothed'] = averaged_asterix_aper_df['y'].rolling(window=window, center=True).mean()

# Plot the results
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(averaged_asterix_er_df['x'], averaged_asterix_er_df['y_smoothed'], label='ER', linewidth=2)
plt.plot(averaged_asterix_per_df['x'], averaged_asterix_per_df['y_smoothed'], label='PER', linewidth=2)
plt.plot(averaged_asterix_aper_df['x'], averaged_asterix_aper_df['y_smoothed'], label='APER', linewidth=2)

# Add labels and title
plt.xlabel('Frames', fontsize=12)
plt.ylabel('Reward', fontsize=12)
plt.title('Evaluation on Asterix Env (window=50)', fontsize=14)

# Add a legend
plt.legend()

# Add gridlines for better visualization
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot in the notebook
plt.show()
```
